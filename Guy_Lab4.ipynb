Start by importing 'rasterio' and 'matplotlib' packages:

```python
import rasterio
from rasterio import plot
from rasterio.merge import merge
import glob
import os
import matplotlib.pyplot as plt
```

Use 'rasterio' package to open both images:

```python
image1 = rasterio.open("image1.tif")
image2 = rasterio.open("image2.tif")
```

```python
image1_bands = str(image1.count)
image2_bands = str(image2.count)
print("Image 1 has " + image1_bands + " bands and Image 2 has " + image2_bands + " bands.")
```

```python
plot.show(image1, cmap='terrain')
```

```python
plot.show(image2, cmap='terrain')
```

```python
image1_crs = str(image1.crs)
image2_crs = str(image2.crs)
print("Image 1 CRS: " + image1_crs)
print("Image 2 CRS: " + image2_crs)
```

Thus, both images have a CRS of 32618, or WGS 84 / UTM zone 18N.

The following code plots the 3 bands of Image 1, which are the red, green, and blue bands:

```python
fig, (axr, axg, axb) = plt.subplots(1,3, figsize=(20,5))
plot.show((image1, 3), ax=axr, cmap='Reds', title='Red Band')
plot.show((image1, 2), ax=axg, cmap='Greens', title='Green Band')
plot.show((image1, 1), ax=axb, cmap='Blues', title='Blue Band')
```

```python
images = [image1, image2]
mosaic, image_mosaic = merge(images)
plot.show(mosaic, cmap='terrain')
```
